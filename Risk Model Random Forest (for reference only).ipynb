```python

def YearFreq(dff):
    i = 0
    j = 0
    while i != dff.shape[0]:
        yearfreq = 0
        userid = dff['UserId'][i]
        firstday = dff['TransDate'][i]
        yearfreq = yearfreq + 1
        days_in_a_year = 365
        j = j + 1
        
        while (j != dff.shape[0]) and (dff['UserId'][j] == userid):
            if (dff['TransDate'][j] - firstday).days < days_in_a_year:
                yearfreq = yearfreq + 1
            else:
                break
                
            if (j == dff.shape[0]):
                pass
            else:
                j = j + 1
                
        while i != j:
            
            dff['YearFrequency'][i] = yearfreq
            i = i + 1
        
            
```

```python
def MonthFreq(dff):
    i = 0
    j = 0
    while i != dff.shape[0]:
        monthfreq = 0
        userid = dff['UserId'][i]
        firstday = dff['TransDate'][i]
        monthfreq = monthfreq + 1
        days_in_a_month = 30
        j = j + 1
        
        while (j != dff.shape[0]) and (dff['UserId'][j] == userid):
            if (dff['TransDate'][j] - firstday).days < days_in_a_month:
                monthfreq = monthfreq + 1
            else:
                break
                
            if (j == dff.shape[0]):
                pass
            else:
                j = j + 1
                
        while i != j:
            
            dff['MonthFrequency'][i] = monthfreq
            i = i + 1
                
```

```python
def WeekFreq(dff):
    i = 0
    j = 0
    while i != dff.shape[0]:
        weekfreq = 0
        userid = dff['UserId'][i]
        firstday = dff['TransDate'][i]
        weekfreq = weekfreq + 1
        days_in_a_week = 7
        j = j + 1
        
        while (j != dff.shape[0]) and (dff['UserId'][j] == userid):
            if (dff['TransDate'][j] - firstday).days < days_in_a_week:
                weekfreq = weekfreq + 1
            else:
                break
                
            if (j == dff.shape[0]):
                pass
            else:
                j = j + 1
                
        while i != j:
            
            dff['WeekFrequency'][i] = weekfreq
            i = i + 1
                
```

```python
def FreqEachYear(dff):
    i = 0
    j = 0
    
    
    while i != dff.shape[0] :
        Freq2017 = 0
        Freq2018 = 0
        Freq2019 = 0
        userid = dff['UserId'][i]
        
        while (j != dff.shape[0]) and (dff['UserId'][j] == userid) :
            if dff['TransDate'][j].year == 2017:
                Freq2017 = Freq2017 + 1
            elif dff['TransDate'][j].year == 2018:
                Freq2018 = Freq2018 + 1
            elif dff['TransDate'][j].year == 2019:
                Freq2019 = Freq2019 + 1
            if (j == dff.shape[0]):
                pass
            else:
                j = j + 1
        #userid = dff['UserId'][j]
        while i != j:
            if (dff['TransDate'][i].year == 2017) and (dff['UserId'][i] == userid) :
                dff['Freq2017'][i] = Freq2017
           
            elif (dff['TransDate'][i].year == 2018) and (dff['UserId'][i] == userid) :
                dff['Freq2018'][i] = Freq2018
            
            elif (dff['TransDate'][i].year == 2019) and (dff['UserId'][i] == userid) :
                dff['Freq2019'][i] = Freq2019
            i = i + 1
       
        
            
        
```

```python
X = df.drop(['UserId', 'TransDate', 'Source of Fund Others', 'TransactionStatus', 
             'IssuerCode', 'RiskScore'], axis = 1) 
```

```python
x_high=X[X['Risk Ranking']=='High']
x_low=X[X['Risk Ranking']=='Low']
x_medium=X[X['Risk Ranking']=='Medium']
```

```python
from sklearn.utils import resample

print("Shape of High Risk : ",x_high.shape)
print("Shape of Medium Risk : ",x_medium.shape)
print("Shape of Low Risk : ",x_low.shape)
```

```python
x_high_unsampled=resample(x_high, replace=True,n_samples=40000,random_state=42)
x_low_unsampled=resample(x_low, replace=False,n_samples=50000,random_state=42)
x_medium_unsampled=resample(x_medium,replace=True,n_samples=40000,random_state=42)
x_low_unsampled.head()
```

```python
x_sampled=pd.concat([x_medium,x_high_unsampled,x_low_unsampled])
x_sampled['Risk Ranking'].value_counts()

X=shuffle(x_sampled)
Z = X.drop(['Risk Ranking'], axis = 1)
```

```python
y = X['Risk Ranking']
x = X.drop(['Risk Ranking'], axis = 1)
x = pd.get_dummies(x)
x.head()
```

```python
import pickle

pickle_filename = "Z.pkl"  

with open(pickle_filename, 'wb') as file:  
    pickle.dump(Z, file)
    
pickle_filename = "x.pkl"  

with open(pickle_filename, 'wb') as file:  
    pickle.dump(x, file)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators = 1200, max_features = 'sqrt', max_depth = 180)
model.fit(X_train, y_train)

model_predict = model.predict(X_test)
```

```python
model_cv_score = cross_val_score(model, x, y, cv = 10)

print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, model_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, model_predict))
print('\n')
print("=== All AUC Scores ===")
print(model_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", model_cv_score.mean())
```

```python
feature_list = x.columns

# Get numerical feature importances
importances = list(model.feature_importances_)

# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]

# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
